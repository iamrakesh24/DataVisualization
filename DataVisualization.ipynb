```python
x=np.arange(10)
y=x**2
z=x+y
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(10,10)) #figure size is used to give the size of the graph 
plt.plot(x,y,color="red",label="A",linestyle="dashed",marker='*')#plot graph between x and y
plt.plot(x,z,color='blue',label="B")
plt.xlabel("time") # gives name to the x axis
plt.ylabel("skill") # gives name to the y axis
plt.title("skills over time") # gives a title to the graph

plt.legend()
plt.show()
```

### Bar Graph

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.bar([1,2,5],[10,20,15]) #takes two arguements x-axis and y-axis
```

```python
coordinates=np.array([0,1,2])*2
plt.bar(coordinates,[5,8,12])
plt.bar(coordinates,[10,15,18])
plt.show()
```

```python
plt.style.use("dark_background")
coordinate=np.array([0,1,2])*2
plt.bar(coordinate+0.5,[10,15,20],width=0.5,label="current price",tick_label=["Gold","silver","diamond"],color="blue")
plt.bar(coordinate-0.5,[17,20,12],width=0.5,label="previous price",color="orange")

plt.xlabel("2020")
plt.ylabel("2021")

plt.title("PriceComparison")
plt.legend()
plt.show()
```

### Pie Charts

```python
subjects="OS","DBMS","C++","SQL"
marks=[100,54,86,36]
```

```python
plt.pie(marks,labels=subjects,explode=(0,0,0.1,0)) #explode is used to make a particular section of the pie chart pop out
plt.title("Semester")
plt.show()
```

### Histogram

```python
xsn=np.random.randn(100)
sigma=7
u=50
x=np.round(xsn*sigma+u)
u=np.round(xsn*15+40)
print(x)
```

```python
plt.style.use("seaborn")
plt.hist(x,alpha=0.8) #alpha is used to show the opacity content in case the two histograms overlap each other
plt.hist(u,alpha=0.8)
plt.xlabel("Marks")
plt.ylabel("FrequencyCount/Probability")
plt.show()
```
